```python
# Implement your solution here
from transformers import AutoTokenizer, AutoModel

# load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained('allenai/specter')
model = AutoModel.from_pretrained('allenai/specter', device_map= 'cuda:0')
```

```python
tabs = pd.read_csv('title_abs.csv')
```

```python
tabs['embed'] = tabs['title_abs'].progress_apply(lambda x: model(**tokenizer(x, padding=True, truncation=True, return_tensors="pt",
                             max_length=512).to('cuda:0')).last_hidden_state[:,0,:].detach().cpu().numpy())
```
